```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # registers 3D projection
```

```python
file_name = "FUN.DTLZ3.Evolver.csv"
#file_name = "FUN.DTLZ3.OriginalNSGAII.csv"
data = pd.read_csv(file_name, header=None)
```

```python
# Create 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(data[0], data[1], data[2], color="black", s=40)

# Labels
ax.set_xlabel("f1")
ax.set_ylabel("f2")
ax.set_zlabel("f3")

# Set initial viewing angle (optional, user can still rotate interactively)
ax.view_init(elev=20, azim=45)

ax.set_title("DTLZ3. NSGA-II (Evolver)")
```
